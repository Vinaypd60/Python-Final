# List category Data-Type(Collection DT)
- The purpose of List Category DT is that "to store multiple values either same type (or) different type (or) both the types" in single objects with unique and duplicate.
- We have two Data-types in List category
    - 1) List (Mutable)
    - 2) Tuple (Immutable)

# List[]:
- List is one of the pre-defined class and treated as list category DT.
- The purpose of List Category DT is that "to store multiple values either same type (or) different type (or) both the types" in single objects with unique and duplicate.
- To store list data, we use square brackets.
    - syntax: varname = [val1, val2, ...., valn]
- Here varname is an object of list

```python
lst = [10,20,30,40,20,10]
print(lst,type(lst),id(lst))
```

```python
lst = [10,"vinay","kritika",2+3j,2.45,10]
print(lst,type(lst),id(lst))
```

- An object of list maintains insertion order

```python
lst[0]
```

```python
lst[1]
```

```python
lst[4]
```

- An object of list belongs to mutable because list object allows us to perform modification / updation / assignment by using indexing and slicing.

# Pre-defined functions in list:
- We know that the object of list, we can perform both indexing and slicing operations.
- With indexing, we get / access values and we can update single value of list object.
- With slicing operation we get / access multiple values and we can update multiple value  of list object.
- To do more operations on list object along with index and slicing operation, we use the pre-defined functions present in list.
    - 1) append()
    - 2) insert()
    - 3) remove()
    - 4) pop(index)
    - 5) pop()
       - del operator
    - 6) clear()
    - 7) index()
        - enumerator
    - 8) copy()
    - 9) count()
    - 10) reverse()
    - 11) sort()
    - 12) extend()

# 1) append()
- This function is used for adding the element / values at the end of the list.
    - syntax: listobj.append(values)

```python
lst = []
print(lst,type(lst),id(lst))
print("------------------------")
lst.append(10)
print(lst)
```

```python
lst.append(20)
print(lst)
```

```python
lst.append("vinay")
print(lst)
```

```python
lst.append(2+3j)
print(lst)
```

```python
lst.append(1.5)
print(lst)
```

# 2) insert()
- This function is used for inserting an element in the existing list at specified index.
  - Here index can be either positive or Negative

```python
lst = [10,20,"vinay",2+4j]
print(lst,type(lst),id(lst))
print("---------------------------------")
lst.insert(3,3.6)
print(lst,type(lst))
```

# 3) remove()
- This function is used for removing the specified val from list object.
- If the specified value does not exist then we get error.
        - lstobj.remove(val)

```python
lst = [10,20,"Vinay","Kritika",1.4,2+5j]
print(lst,type(lst),id(lst))
print("-----------------------------------")
lst.remove(20)
print(lst,type(lst),id(lst))
```

# 4) pop(index) --> Based on index removing
- This function is used for removing an element of list based on valid index.
- Index can be either positvie or negative
    - syntax: listobj.pop(ind)

```python
lst = [10,20,"Vinay","Kritika",1.4,2+5j]
print(lst,type(lst),id(lst))
print("-----------------------------------")
lst.pop(4)
print(lst,type(lst),id(lst))
```

```python
lst = [10,20,"Vinay","Kritika",1.4,2+5j]
print(lst,type(lst),id(lst))
print("-----------------------------------")
lst.pop(1)
print(lst,type(lst),id(lst))
```

# 5) pop() --> LIFO
- This function is used for removing the last element of list.
    - Syntax: listobj.pop() ---> Internally -1 will take

```python
lst = [10,20,"Vinay","Kritika",1.4,2+5j]
print(lst,type(lst),id(lst))
print("-----------------------------------")
lst.pop()
print(lst,type(lst),id(lst))
```

# del operator
- Syntax: del listobj[index] ---> Removing one element
- Syntax: del listobj[Begin index : End index] ---> Removes element

```python
lst = [10,20,"Vinay","Kritika",1.4,2+5j]
print(lst,type(lst),id(lst))
print("-----------------------------------")
del lst[-2]
print(lst)
```

```python
lst = [10,20,"Vinay","Kritika",1.4,2+5j]
print(lst,type(lst),id(lst))
print("-----------------------------------")
del lst[2:4]
print(lst,type(lst),id(lst))
```

# 7) clear()
- This function is used for removing all elements.
- If we call this function upon empty list object then we get space (or) None as result.

```python
lst = [10,20,"Vinay","Kritika",1.4,2+5j]
print(lst,type(lst),id(lst))
print("-----------------------------------")
lst.clear()
print(lst,type(lst),id(lst))
```

# 8) Set() --> It will removes duplicates

```python
lst = [10,20,"Vinay","Kritika",1.4,2+5j,10,20,"Sai"]
print(lst,type(lst),id(lst))
print("-----------------------------------")
s = set(lst)
print(s,type(s),id(s))
```

# 9) Index()
- This function is used for obtaining the index of 1st occurences of specified element.
- If the specified values does not exist then we will get value error.
    - Syntax: listobj.index(val)

```python
lst = [10,20,"Vinay","Kritika",1.4,2+5j,10,20]
print(lst,type(lst),id(lst))
print("-----------------------------------")
lst.index(10)
```

# count()

```python
lst = [10,20,"Vinay","Kritika",1.4,2+5j,10,20]
print(lst,type(lst),id(lst))
print("-----------------------------------")
lst.count(10)
```

# reverse()

```python
lst = [10,20,"Vinay","Kritika",1.4,2+5j]
print(lst,type(lst),id(lst))
print("-----------------------------------")
lst.reverse()
print(lst,type(lst),id(lst))
```

# sort()

```python
lst = [10,20,34,45,11,67,30]
print(lst,type(lst),id(lst))
print("-----------------------------------")
lst.sort()
print(lst,type(lst),id(lst))
```

# extend()

```python
lst = [10,20,30]
lst1 = [40,50,60]
print(lst,type(lst),id(lst))
print("-------------------------")
print(lst1,type(lst1),id(lst1))
print("--------------------------")
lst.extend(lst1)
print(lst)
```
