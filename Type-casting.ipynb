- The process of converting one possible type of value into another possible type of value is called "Type casting".
- In python programming, we have 5-fundamental type casting techniques.
    - 1) Int()
    - 2) Float()
    - 3) bool()
    - 4) complex()
    - 5) str()

# 1) Int():
- This function is used for converting one type of possible value into int type value.
    - Syntax: varname = int(float / bool / comp / str)

```python
a = 12.34
print(a,type(a),id(a))
print("--------------------------")
b = int(a)
print(b,type(b),id(b))
```

```python
a = 23.5
print(a,type(a),id(a))
print("-------------------------")
b = int(a)
print(b,type(b),id(b))
```

# boolean to int:

```python
a = True
print(a,type(a),id(a))
print("------------------------")
b = int(a)
print(b,type(b),id(b))
```

```python
x = False
print(x,type(x),id(x))
print("------------------------")
y = int(x)
print(y,type(y),id(y))
```

# complex to int: not possible

# str to int: not possible

# 2) Float():
- This function is used for converting one type of possible value into another type of possible value.
    - syntax: varname = float(int / bool / complex / str)

# int to float:

```python
a = 14
print(a,type(a),id(a))
print("----------------------")
b = float(a)
print(b,type(b),id(b))
```

```python
a = 36
print(a,type(a),id(a))
print("----------------------")
b = float(a)
print(b,type(b),id(b))
```

# bool to float: posible

# complet to float: not possible

# str float to float: possible

# str bool to float: not possible

# 3) bool():
- This function is used for converting one type of possible value to another type of possible value.
    - syntax: varname = bool(int / float / complex / str)

# int to bool

```python
a = 100
print(a,type(a),id(a))
print("----------------------")
b = bool(a)
print(b,type(b),id(b))
```

```python
a = -123
print(a,type(a),id(a))
print("-----------------------")
b = bool(a)
print(b,type(b),id(b))
```

# float to bool

```python
a = 2.5
print(a,type(a),id(a))
print("-----------------------")
b = bool(a)
print(b,type(b),id(b))
```

```python
a = 0.0
print(a,type(a),id(a))
print("-----------------------")
b = bool(a)
print(b,type(b),id(b))
```

 # complex to bool

```python
a = 2 + 5j
print(a,type(a),id(a))
print("-----------------------")
b = bool(a)
print(b,type(b),id(b))
```

```python
a = 0 + 0j
print(a,type(a),id(a))
print("-----------------------")
b = bool(a)
print(b,type(b),id(b))
```

- All Non-Zero values are treated as True
- All Zero values are treated as False
