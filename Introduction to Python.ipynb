# Variables:
      - A variable is an identifiers, whose values can be changes during execution of the program.
      - In python variables are called 'Objects'.
### Rules of using variables:
      - Var names is combination of alphabet, digits and symbols & underscore.
      - Var name must start with alphabet (or) _
      - Within the var name, special symbols are not allowed
      - All var names are case sensitive
      - Choose the var name such way that they must be user-friendly
      - All pre-defined classNames (Int, float) we can use
