# Int:
- Int is one of the predefined class and it is treated as "Fundamental categorical DT".
- The purpose of int DT is that to store integers / Whole / integral values.
- Integers / Whole / integral values are nothing but without decimal values.

```python
a = 123 # Here a is variable and 123 are literals
print(a,type(a),id(a))
```

```python
b = 234
print(b,type(b),id(b))
```

# Sequence category DT:
- The purpose of sequence category DT is that to store sequence of values.
- In python programming, we have 4 - DT in sequence category.
- The purpose of 'str' DT is that to store string \ text \ alphanumeric data.
- Number of ways to organize / store str data.
- In python programming, we have two ways to organize str data.
    - 1) single line str data
    - 2) multi line str data

# 1) Single line / str Data
- The single line string data is one, which is enclosed with single quotes (or) double quotes.
- syntax: "single" or 'single'

```python
s = "vinay"
print(s,type(s),id(s))
```

```python
s1 = "12345"
print(s1,id(s1),type(s1))
```

```python
s2 = "vinay1234"
print(s2,type(s2),id(s2))
```

```python
s3 = "vinay123@#$"
print(s3,type(s3),id(s3))
```

- we can't use single quotes and double quotes for organizing multiline str data.
- A string is a collection of char enclosed with '' or "" quotes.

# Multi-line str data
- The multiline string data is one, which is enclosed with tripple single or double quotes.

```python
address = """ vinay is a good boy
              he is from hyderabad, telangana
              H.no 576/D narendra nagar colony
              India """
print(address)
```
